This Jupyter notebook includes some code to get you started with web scraping. We will use a package called `BeautifulSoup` to collect the data from the web. Once you've collected your data and saved it into a local `.csv` file you should start with your analysis.

### Scraping data from Skytrax

If you visit [https://www.airlinequality.com] you can see that there is a lot of data there. For this task, we are only interested in reviews related to British Airways and the Airline itself.

If you navigate to this link: [https://www.airlinequality.com/airline-reviews/british-airways] you will see this data. Now, we can use `Python` and `BeautifulSoup` to collect all the links to the reviews and then to collect the text data on each of the individual review links.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 34
page_size = 100

reviews = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')

    aircraft=travel_type=seat=date=recommend=""
    seat_comfort=cabin_rev=food_rev=ground_rev=money_rev=0
    for para in parsed_content.find_all("article", {"itemprop": "review"}):
        # reviews.append(para.get_text())
        review=para.find("div",{"class":"text_content"}).get_text()
        # print(para.prettify())
        # print(int(para.find("span",{"itemprop":"ratingValue"}).get_text()))
        # print(para.find("div", {"class": "text_content"}).get_text())
        # print(len(para.find_all("td",{"class":"review-value"})))
        # print(len(para.find_all("tr")))
        for td in para.find_all("tr"):
            # print(td.prettify())
            # print(td.contents[0].get_text())
            if td.contents[0].get_text()=="Aircraft":
                aircraft=td.contents[1].get_text()
            if td.contents[0].get_text()=="Type Of Traveller":
                travel_type=td.contents[1].get_text()
            if td.contents[0].get_text()=="Seat Type":
                seat=td.contents[1].get_text()
            if td.contents[0].get_text()=="Date Flown":
                date=td.contents[1].get_text()
            if td.contents[1].get_text()=="Seat Comfort":
                seat_comfort=len(td.find_all("span", {"class","star fill"}))
            if td.contents[1].get_text()=="Cabin Staff Service":
                cabin_rev=len(td.find_all("span", {"class","star fill"}))
            if td.contents[1].get_text()=="Food & Beverages":
                food_rev=len(td.find_all("span", {"class","star fill"}))
            if td.contents[1].get_text()=="Ground Service":
                ground_rev=len(td.find_all("span", {"class","star fill"}))
            if td.contents[1].get_text()=="Value For Money":
                money_rev=len(td.find_all("span", {"class","star fill"}))
            if td.contents[0].get_text()=="Recommended":
                recommend=td.contents[1].get_text()
            
            # print(td.contents)
        # print(aircraft,travel_type,seat,date)
        reviews.append((review,aircraft,travel_type,seat,date,seat_comfort,cabin_rev,food_rev,ground_rev,money_rev,recommend))

    
    print(f"   ---> {len(reviews)} total reviews")
```

```python
df = pd.DataFrame(reviews, columns=['Review','Aircraft', 'Travel Type', 'Seat Type', 'Date Flown',"Seat Comfort","Cabin Staff Service","Food & Beverages","Ground Service","Value For Money","Recommendation"]) 
df.head()
```

```python
df.to_csv("data/BA_reviews_depth.csv")
```

With this method we are able collected any specified number we want, in this case we have collected 3400 reviews by iterating through the paginated pages on the website.
